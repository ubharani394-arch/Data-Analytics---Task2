```python

import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_excel('/content/ApexPlanet_DataAnalytics_Dataset.xlsx')

# First 5 Rows
print(df.head())

# Dataset Info
print(df.info())

# Statistical Summary
print(df.describe())

# Mean, Median, Mode of Age
print("Mean Age:", df['Age'].mean())
print("Median Age:", df['Age'].median())
print("Mode Age:", df['Age'].mode()[0])

# Total Sales
print("Total Sales:", df['Total_Sales'].sum())

# Top 5 Cities by Sales
print(df.groupby('City')['Total_Sales'].sum().sort_values(ascending=False).head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
df['Age'].hist(bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
category_sales = df.groupby('Category')['Total_Sales'].sum()

category_sales.plot(kind='bar')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
df['Category'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)
plt.title('Category Distribution')
plt.ylabel('')
plt.show()
```
